```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/time_series_2019-ncov-Confirmed.csv')
df
```

```python
for i in range(4,487):
    count = 0
    x = np.array(df.iloc[:,i])
    for j in range(len(x)):
        if x[j] == 0:
            count += 1
    if (count/487)*100 > 70:
        del df[df.columns[i]]
        
        
```

```python
del df['Province/State']
del df['Lat']
del df['Long']
df
```

```python
df1 = df.groupby('Country/Region').sum()
df1
```

```python
plt.figure(figsize=(10,6))
x1 = np.array(df1.index[35:70])
y1 = np.array(df1.iloc[:,5][35:70])
plt.bar(x1,y1)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('No. of cases')
plt.title('1. Country vs No of cases on 03/09/20')
plt.show()
```
