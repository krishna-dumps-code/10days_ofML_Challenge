```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/train.csv')
```

```python
del df['PassengerId']
del df['Name']
del df['Ticket']
```

```python
sex = np.array(df.iloc[:,2])
for i in range(len(sex)):
    if sex[i] == 'male':
        sex[i] = 0
    else:
        sex[i] = 1
df.iloc[:,2] = sex
```

```python
emb = np.array(df.iloc[:,8])
for i in range(len(emb)):
    if emb[i] == 'S':
        emb[i] = 0
    elif emb[i] == 'C':
        emb[i] = 1
    else:
        emb[i] = 2
df.iloc[:,8] = emb
```

```python
df1 = df.groupby('Sex').sum()
```

```python
del df['Cabin']
```

```python
age = np.array(df.iloc[:,3])
for i in range(len(age)):
    if age[i] <= 15:
        age[i] = 0
    elif age[i] > 15 and age[i] <= 38:
        age[i] = 1
    elif age[i] > 38 and age[i] <= 50:
        age[i] = 2
    else:
        age[i] = 3
df['AgeGroup'] = age
del df['Age']
```

```python
fare = np.array(df.iloc[:,5])
for i in range(len(fare)):
    if fare[i] <= 50:
        fare[i] = 0
    elif fare[i] > 50 and fare[i] <= 100:
        fare[i] = 1
    elif fare[i] > 100 and fare[i] <= 300:
        fare[i] = 2
    else:
        fare[i] = 3
df['Fare_range'] = fare
del df['Fare']
```

```python
comp = np.array(df.iloc[:,4])
for i in range(len(comp)):
    if comp[i] > 0 and comp[i] < 3:
        comp[i] = 1
    elif comp[i] > 2:
        comp[i] = 2
        
df['Companion_range'] = comp
```

```python
df2 = df.groupby('AgeGroup').sum()
df3 = df.groupby('Fare_range').sum()
df4 = df.groupby('Companion_range').sum()
```

```python
fig = plt.figure(figsize=(12,4))
a1 = np.array(['Children','Teenage & Youth','Men/Women','Old'])
b1 = np.array(df2.iloc[:,0])
a2 = np.array(['Upto 50','50-100','100-300','>300'])
b2 = np.array(df3.iloc[:,0])
a3 = np.array(['0','1 or 2','3 or more'])
b3 = np.array(df4.iloc[:,0])
plt.subplot(1,3,1)
plt.bar(a1,b1, color='g')
plt.ylabel('Survival')
plt.xlabel('AgeGroup')
plt.xticks(rotation=90)
plt.subplot(1,3,2)
plt.bar(a2,b2, color='r')
plt.ylabel('Survival')
plt.xlabel('Fare_range')
plt.xticks(rotation=90)
plt.subplot(1,3,3)
plt.bar(a3,b3, color='m')
plt.ylabel('Survival')
plt.xlabel('companions with Passenger')
plt.xticks(rotation=90)
fig.tight_layout()
plt.show()
```
