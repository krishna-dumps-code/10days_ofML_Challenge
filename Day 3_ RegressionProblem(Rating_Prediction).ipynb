```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('http://iali.in/datasets/cereal.csv')
df[29:47]
```

```python
del df['name']
del df['type']
del df['shelf']
del df['weight']
del df['cups']
```

```python
df
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['mfr'] = le.fit_transform(df['mfr'])
df['vitamins'] = le.fit_transform(df['vitamins'])
df['sugars'] = le.fit_transform(df['sugars'])
df['potass'] = le.fit_transform(df['potass'])
```

```python
features = ['mfr','calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']
X = df[features]
y = df['rating']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(X_train,y_train)
y_pred = lreg.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dreg = DecisionTreeRegressor()
dreg.fit(X_train,y_train)
y_pred = dreg.predict(X_test)
```

```python
from sklearn import metrics
print('Accuracy of Linear Regression model: ',lreg.score(X_test,y_test))
print('Accuracy of Decision tree Regression model: ',dreg.score(X_test,y_test))
```
